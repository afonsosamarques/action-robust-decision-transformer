```python
def count_files_and_dirs(root_path):
    for root, dirs, files in os.walk(root_path):
        num_dirs = len(dirs)
        num_files = len(files)
        print(f"Directory: {root.split('/')[-1]}")
        print(f"  Number of subdirectories: {num_dirs}")
        print(f"  Number of files: {num_files}")


def count_files_and_dirs_one_level(root_path):
    for dir_name in os.listdir(root_path):
        dir_path = os.path.join(root_path, dir_name)
        
        if os.path.isdir(dir_path):
            total_files = 0
            total_dirs = 0
            
            for _, dirs, files in os.walk(dir_path):
                total_files += len(files)
                total_dirs += len(dirs)
                
            print(f"Directory: {dir_path.split('/')[-1]}")
            print(f"  Total number of subdirectories: {total_dirs}")
            print(f"  Total number of files: {total_files}")


def count_json_files_and_dirs(root_path):
    l = []
    nl = []
    d_files = {}
    d_dirs = {}
    for first_level_dir in os.listdir(root_path):
        first_level_dir_path = os.path.join(root_path, first_level_dir)
        
        if os.path.isdir(first_level_dir_path):
            for second_level_dir in os.listdir(first_level_dir_path):
                second_level_dir_path = os.path.join(first_level_dir_path, second_level_dir)
                
                if os.path.isdir(second_level_dir_path):
                    total_files = 0
                    total_dirs = 0
                    
                    for _, dirs, files in os.walk(second_level_dir_path):
                        total_files += len([f for f in files if f.endswith('.json')])

                        valid_dirs = []
                        for dir in dirs:
                            if dir.startswith("env-adv"): valid_dirs.append(dir)
                        total_dirs += len(valid_dirs)
                        

                        if len(valid_dirs) < 2:
                            name = second_level_dir_path.split('/')[-1]
                            dataset = name.split('-')[1] if "-" in name else name
                            l.append(dataset)
                        else:
                            name = second_level_dir_path.split('/')[-1]
                            dataset = name.split('-')[1] if "-" in name else name
                            nl.append(dataset)
                    
                    # add to dictionary model name and number of files
                    d_dirs[second_level_dir_path.split('/')[-1]] = total_dirs
                    d_files[second_level_dir_path.split('/')[-1]] = total_files

    return d_dirs, d_files, l, nl
    

# Initialize root directory path
root_path = os.path.join(os.getcwd(), "results-thesis")

# Call the function to count files and directories
d_dirs, d_files, l, nl = count_json_files_and_dirs(root_path)

# print(len(set(l)))
# print(len(set(nl)))
# print(set(l) - set(nl))
# print(set(nl) - set(l))

print("\nFile Count:")
d_files = dict(sorted(d_files.items(), key=lambda item: item[1], reverse=True))
for k, v in d_files.items():
    print(f"{k}: {v}")
```
